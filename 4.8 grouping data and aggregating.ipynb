# create a column for frequent and non frequent customers

```python
ords_prods_merged[['user_id', 'loyalty_flag', 'days_since_prior_order']].head(60)
```

```python
ords_prods_merged.groupby('loyalty_flag').agg({'days_since_prior_order': ['mean', 'min', 'max', 'median']})
```

```python
ords_prods_merged['buying_freq'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merged.head(20)
```

```python
ords_prods_merged.loc[ords_prods_merged['buying_freq'] > 20, 'customer_freq'] = 'Non-frequent customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['buying_freq'] > 10) & (ords_prods_merged['buying_freq'] <= 20), 'customer_freq'] = 'regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['buying_freq'] <= 10, 'customer_freq'] = 'frequent customer'
```
