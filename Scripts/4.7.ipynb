# creating conditions/flags that sorts products by price

```python
#define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#using the function
df['price_range'] = df.apply(price_label, axis=1)
```

# task 4.7 Create a column for busiest days and slowest days

```python
#question 2
```

```python
new_result = []

for value in df_ords_merged["orders_day_of_week"]:
  if value == 0:
    new_result.append("Busiest_days")
  elif value == 1:
    new_result.append("Busiest_days")
  elif value == 3:
    new_result.append("Slowest_days")
  elif value ==4:
    new_result.append("Slowest_days")
  else:
    new_result.append("Regularly busy")
```

```python
new_result
```

```python
df_ords_merged['Busiest day'] = new_result
```

```python
df_ords_merged['Busiest day'].value_counts(dropna = False)
```

```python
#question 4 create new column for busiest periods of the day 
```

```python
updated_result = []

for value in df_ords_merged["order_hour_of_day"]:
  if value > 10 & value <=18:
    updated_result.append("Most_orders")
  elif value <=5:
    updated_result.append("Fewest_orders")
  else:
    updated_result.append("Average_orders")
  
```

```python
df_ords_merged['Busiest Hours'] = updated_result
```
